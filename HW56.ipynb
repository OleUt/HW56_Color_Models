```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as clr
%matplotlib inline
```

```python
def to_print(fig, title, cmap):
    plt.axis('off')
    plt.imshow(fig, cmap=cmap)
    plt.title(title)
```

```python
dataroot = 'C:/Users/MyWork/PycharmProjects/HW56_Color_Models/cat.jpg'
```

```python
image = cv2.imread(dataroot)
to_print(image,'image_BGR', None)
```

```python
# RGB

image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

fig = plt.figure(figsize=(10, 8))

fig.add_subplot(2, 2, 1)
to_print(image_RGB,'image_RGB', None)

# RGB red
image_RGB_red = image_RGB.copy()
image_RGB_red[:,:,(1,2)] = 0
fig.add_subplot(2, 2, 2)
to_print(image_RGB_red,'image_RGB_red', None)

# RGB green
image_RGB_green = image_RGB.copy()
image_RGB_green[:,:,(0,2)] = 0
fig.add_subplot(2, 2, 3)
to_print(image_RGB_green,'image_RGB_green', None)

# RGB blue
image_RGB_blue = image_RGB.copy()
image_RGB_blue[:,:,(0,1)] = 0
fig.add_subplot(2, 2, 4)
to_print(image_RGB_blue,'image_RGB_blue', None)

plt.show(fig)
```

```python
# HSV

image_HSV = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

fig = plt.figure(figsize=(10, 28))
fig.add_subplot(6, 2, 1)
to_print(image_HSV,'image_HSV', None)

# HSV -H

image_HSV_H = image_HSV.copy()
#  S  V   (255) 
image_HSV_H[:, :, (1,2)] = 255
fig.add_subplot(6, 2, 3)
to_print(image_HSV_H,'image_HSV_H', None)

#   RGB
image_HSV_H_RGB =  cv2.cvtColor(image_HSV_H, cv2.COLOR_HSV2RGB) 
fig.add_subplot(6, 2, 4)
to_print(image_HSV_H_RGB,'image_HSV_H_RGB', None)

# HSV -S

image_HSV_S = image_HSV.copy()
#  H  (179)
image_HSV_S [:, :, 0] = 179 
#  V  (255)
image_HSV_S [:, :, 2] = 255
fig.add_subplot(6, 2, 5)
to_print(image_HSV_S,'image_HSV_S', None)

#   RGB
image_HSV_S_RGB =  cv2.cvtColor(image_HSV_S, cv2.COLOR_HSV2RGB) 
fig.add_subplot(6, 2, 6)
to_print(image_HSV_S_RGB,'image_HSV_S converted to RGB', None)

# HSV -V  (1)

image_HSV_V = image_HSV.copy()
#  H  (179)
image_HSV_V[:, :, 0] = 179
#  S  0
image_HSV_V[:, :, 1] = 0
fig.add_subplot(6, 2, 7)
to_print(image_HSV_V,'image_HSV_V', None)

#   RGB
image_HSV_V_RGB =  cv2.cvtColor(image_HSV_V, cv2.COLOR_HSV2RGB) 
fig.add_subplot(6, 2, 8)
to_print(image_HSV_V_RGB,'image_HSV_V converted to RGB', None)

# HSV -V  (2)

image_HSV_V = image_HSV.copy()
#  H  (179)
image_HSV_V[:, :, 0] = 179
#  S  (225)
image_HSV_V[:, :, 1] = 255
fig.add_subplot(6, 2, 9)
to_print(image_HSV_V,'image_HSV_V', None)

#   RGB
image_HSV_V_RGB =  cv2.cvtColor(image_HSV_V, cv2.COLOR_HSV2RGB) 
fig.add_subplot(6, 2, 10)
to_print(image_HSV_V_RGB,'image_HSV_V converted to RGB', None)

plt.show(fig)
```

```python
# LAB

image_LAB = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

fig = plt.figure(figsize=(10, 18))
fig.add_subplot(4, 2, 1)
to_print(image_LAB,'LAB', None)

# LAB_L

image_LAB_L = image_LAB.copy()

image_LAB_L[:,:, (1,2)] = 0
fig.add_subplot(4, 2, 3)
to_print(image_LAB_L,'LAB_L', None)

# L  -  
image_LAB_L_RGB =  cv2.cvtColor(image_LAB_L,cv2.COLOR_LAB2RGB)
image_LAB_L_GRAY = cv2.cvtColor(image_LAB_L_RGB, cv2.COLOR_RGB2GRAY)
fig.add_subplot(4, 2, 4)
to_print(image_LAB_L_GRAY,'LAB_L converted to RGB, GRAY', 'gray')

# LAB A

a = image_LAB.copy()
a[:, :, 0] = 127
a[:, :, 2] = 0
a_RGB =  cv2.cvtColor(a,cv2.COLOR_LAB2RGB)
a_GRAY = cv2.cvtColor(a_RGB,cv2.COLOR_RGB2GRAY)
fig.add_subplot(4, 2, 5)
to_print(a_GRAY,'LAB_A converted to RGB, GRAY', 'gray')

#     
cmap_a = clr.LinearSegmentedColormap.from_list('custom blue', ['Red', 'White','Green'], N=255)
fig.add_subplot(4, 2, 6)
to_print(a_GRAY,'LAB_A converted to RGB, GRAY, changed colors', cmap_a)

# LAB B

b = image_LAB.copy()
b[:, :, 0] = 127
b[:, :, 1] = 0
b_RGB =  cv2.cvtColor(b,cv2.COLOR_LAB2RGB)
b_GRAY = cv2.cvtColor(b_RGB,cv2.COLOR_RGB2GRAY)
fig.add_subplot(4, 2, 7)
to_print(b_GRAY,'LAB_B converted to RGB, GRAY', 'gray')

#     
cmap_b = clr.LinearSegmentedColormap.from_list('custom blue', ['Yellow',"White",'Blue'], N=255)
fig.add_subplot(4, 2, 8)
to_print(b_GRAY,'LAB_B converted to RGB, GRAY, changed colors', cmap_b)

plt.show(fig)
```
